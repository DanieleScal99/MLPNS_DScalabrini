```python
data=pd.read_csv('DatiRes.csv')
```

```python
data
```

```python
data=data.drop(labels='Unnamed: 1',axis=1)
```

```python
data
```

```python
data['weights']=1/0.01
```

```python
data
```

```python
#ho fittato la line hai dati. Forse ci devo mettere gli errori?
parameters=np.polyfit(data['T'], data['R'], 3, w=data['weights'])
p=np.poly1d(parameters)
x=np.linspace(300,3500) 
```

```python
fig,ax = pl.subplots(1,1,figsize=(9,8))
ax.plot(data['T'].T, data['R'].T, label='Dati sperimentali', marker='.', color = "black", markersize=8, linewidth=0)
ax.plot(x,p(x),label = 'Fit polinomiale n=3 ' ,linestyle='dashed',color='black',alpha=0.4)

ax.set_title('Resistenza del filamento al variare della temperatura')
ax.set_ylabel('R/Ra')
ax.set_xlabel('Temperatura [K]')
ax.set_xlim(200,3600)
ax.legend()

txt=" Andamento con la temperatura della resistenza di un filamento di lampadina. \n I valori di resistenza sono normalizzati a Ra, il valore di resistenza a temperatura ambiente. \nIl fit  stato eseguito con un polinomio di terzo grado."

fig.text(.5, .002, txt, ha='center')

pl.savefig('plotOk.png')
```
