```python
import pandas as pd
import numpy as np
import pylab as pl
```

```python
url="https://bit.ly/videosfile"
```

```python
videos=pd.read_csv(url)
videos
```

```python
videos.columns
```

```python
videos.describe()
```

```python
ncols=videos.describe().columns
videos[ncols].mean()
```

```python
videos[ncols].values #ritorna solo i valori di ncols, storati come array
```

```python
videos_array=videos[ncols].values
videos_array.shape
```

```python
np.median(videos_array, axis = 0)
```

```python
videos.corr() #correlazione tra i dati numerici 
```

```python
pl.imshow(videos[ncols].corr());
pl.colorbar();
pl.xticks([0, 1, 2, 3, 4], ["category_id", "views", "likes", "dislikes", "comment count"], rotation=45);
pl.yticks([0, 1, 2, 3, 4], ["category_id", "views", "likes", "dislikes", "comment count"], rotation=45);
```
